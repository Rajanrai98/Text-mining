```python
## Remove stopwords
```

```python
from nltk.corpus import stopwords
```

```python
my_stop_words = stopwords.words('english')
sw_list = ['\x92' , 'rt','ye','yeah','haha','yes','UOOO1F923' , 'I']
```

```python
my_stop_words.extend(sw_list)

no_stop_tokens = [word for word in text_tokens if not word in my_stop_words]
```

```python
print(no_stop_tokens)
```

```python
## Define the function of the word clouds

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from wordcloud import WordCloud , STOPWORDS
```

```python
def plot_cloud(wordcloud):
    # set figure size
    plt.figure(figsize = (30 , 20))
    # display image
    plt.imshow(wordcloud)
    # no axis details
    plt.axis("off");
```

```python
# Generate wordcloud

stopwords = STOPWORDS
STOPWORDS.add('pron')
STOPWORDS.add('rt')
STOPWORDS.add('yeah')

wordcloud = WordCloud(width = 2000 , height = 2000 , background_color = 'black' , max_words = 100 , stopwords = stopwords).generate(clean_tweets)
plot_cloud(wordcloud)
```
